```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
 
%config Completer.use_jedi = False
%matplotlib inline
```

```python
df = pd.read_csv('No_Outliers_v2.csv', index_col=0)
```

```python
df.T
```

```python
df['Churn'].replace({
    'No': 0,
    'Yes': 1
}, inplace=True)
```

```python
df = df.loc[:,['Age', 'Income', 'Outage_sec_perweek', 'Tenure',
          'MonthlyCharge', 'Bandwidth_GB_Year', 'Population']]
```

```python
df_norm = (df - df.mean()) / df.std()
```

```python
df_norm
```

```python
pca = PCA(n_components=df.shape[1]) # 6 columns 
```

```python
pca.fit(df_norm)
df_pca = pd.DataFrame(pca.transform(df_norm),
                     columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
```

```python
df_pca
```

```python
plt.plot(pca.explained_variance_ratio_ * 100)
plt.xlabel('PCA Components')
plt.ylabel('Explained variance %')
plt.title('Explained Variance Ratio')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.axvline(x=5, c = 'orange', ls= '--')
plt.axvspan(xmin=0, xmax=5, color = 'gray', alpha=.05)
plt.annotate(text='95% of data is explained by 6 components',
             xy=(1,20),
              bbox={"fc": 'gray',
                   'boxstyle': 'round',
                   'alpha': .35,
                   'ec':'b',
                   'lw':2})

# plt.text(1, 20, '95% of data is explained by 6 components',
#          bbox={"fc": 'gray',
#                'boxstyle': 'darrow',
#                'alpha': .35})

plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_ * 100
```

```python
cov_matrix = np.dot(df_norm.T, df_norm) / df.shape[0]
```

```python
cov_matrix
```

```python
eigenvalues = [np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)) for eigenvector in pca.components_]
```

```python
eigenvalues
```

```python
plt.plot(eigenvalues)
plt.xlabel('number of components')
plt.ylabel('eigenvalue')
plt.tight_layout()
plt.show()
```

#  only PC 0-3 are useful because they have eigenvalue of 1 or higher
